## mixup: BEYOND EMPIRICAL RISK MINIMIZATION

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from scipy.stats import beta
```

## 1. mixup of images

```python
im1 = np.array(Image.open("../images/000154.jpg")) 
im2 = np.array(Image.open("../images/000218.jpg")) 
```

```python
plt.figure(num=1, figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(im1)
plt.subplot(1,2,2)
plt.imshow(im2)
plt.show()
```

```python
plt.figure(num=2, figsize=(10,10))

for i in range(1,10):
    lam= i*0.1
    im_mixup = (im1*lam+im2*(1-lam)).astype(np.uint8)
    plt.subplot(3,3,i)
    plt.imshow(im_mixup)
plt.show()
```

## 2. numpy.random.beta(alpha, alpha)

```python
alpha = 1.0
np.random.beta(alpha,alpha)
```

```python
plt.figure(num=3, figsize=(8,5))

x = np.linspace( 0, 1, 100)
a_array = [5, 2, 1, 0.1 ]

for a in a_array:
    plt.plot(x, beta.pdf(x, a, a), lw= 1, alpha= 0.6, label= 'a='+ str(a) + ',b='+ str(a))
    plt.legend(frameon= False)
plt.show()
```
